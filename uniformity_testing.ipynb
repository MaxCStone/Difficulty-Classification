```python
import json
import matplotlib.pyplot as plt
import os
from collections import deque
```

```python
stats_path = "C:\\Users\\stonemc\\OneDrive - Milwaukee School of Engineering\\Documents\\CSC 2621\\Final Project\\Difficulty-Classification\\stats.json"

with open(stats_path, "r") as f:
    stats = json.load(f)

stats_sorted = []

musicxml_dir = "C:\\Users\\stonemc\\OneDrive - Milwaukee School of Engineering\\Documents\\CSC 2621\\Final Project\\Difficulty-Classification\\musicxml"

xmls = []
for file in os.listdir(musicxml_dir):
    if file.endswith(".xml"):
        xmls.append(int(file[0 :-4]))
xmls.sort()

for xml in xmls:
    for stat in stats:
        if stat["filename"] == f"{xml}.xml":
            stats_sorted.append(stat)

distances_left = [stat["distance_k_left"] for stat in stats_sorted]
distances_right = [stat["distance_k_right"] for stat in stats_sorted]
```

```python
def variance(distances):
    mean = sum(distances) / len(distances)
    variance = sum((d - mean) ** 2 for d in distances) / len(distances)
    return variance

variance_left = []
variance_right = []
for dist in distances_left:
    variance_left.append(variance(dist))
for dist in distances_right:
    variance_right.append(variance(dist))
    
greatest_variance_diff = 0
greatest_diff_name = ""

greatest_var_left = 0
greatest_var_right = 0
grestest_var_left_name = ""
grestest_var_right_name = ""
    
for stat in stats_sorted:
    stat["variance_left"] = variance(stat["distance_k_left"])
    stat["variance_right"] = variance(stat["distance_k_right"])
    stat["variance_diff"] = abs(stat["variance_left"] - stat["variance_right"]) 
    if stat["variance_diff"] > greatest_variance_diff:
        greatest_variance_diff = stat["variance_diff"]
        greatest_diff_name = stat["filename"]
    if stat["variance_left"] > greatest_var_left:
        greatest_var_left = stat["variance_left"]
        grestest_var_left_name = stat["filename"]
    if stat["variance_right"] > greatest_var_right:
        greatest_var_right = stat["variance_right"]
        grestest_var_right_name = stat["filename"]
        
print(f"Greatest variance difference: {greatest_variance_diff} in {greatest_diff_name}") 

print(f"Greatest variance left: {greatest_var_left} in {grestest_var_left_name}")
print(f"Greatest variance right: {greatest_var_right} in {grestest_var_right_name}")
    
```
