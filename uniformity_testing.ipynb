```python
import json
import matplotlib.pyplot as plt
import os
from collections import deque
```

```python
stats_path = "C:\\Users\\stonemc\\OneDrive - Milwaukee School of Engineering\\Documents\\CSC 2621\\Final Project\\Difficulty-Classification\\stats.json"

with open(stats_path, "r") as f:
    stats = json.load(f)

stats_sorted = []

musicxml_dir = "C:\\Users\\stonemc\\OneDrive - Milwaukee School of Engineering\\Documents\\CSC 2621\\Final Project\\Difficulty-Classification\\musicxml"

xmls = []
for file in os.listdir(musicxml_dir):
    if file.endswith(".xml"):
        xmls.append(int(file[0 :-4]))
xmls.sort()

for xml in xmls:
    for stat in stats:
        if stat["filename"] == f"{xml}.xml":
            stats_sorted.append(stat)

uniformity_left = [stat["uniformity_left"] for stat in stats_sorted]
uniformity_right = [stat["uniformity_right"] for stat in stats_sorted]
```

```python
greatest_uniformity_left = max(uniformity_left)
greatest_uniformity_right = max(uniformity_right)
smallest_uniformity_left = min(uniformity_left)
smallest_uniformity_right = min(uniformity_right)
print(f"Greatest uniformity left: {greatest_uniformity_left}")
print(f"Smallest uniformity left: {smallest_uniformity_left}")
print(f"Greatest uniformity right: {greatest_uniformity_right}")
print(f"Smallest uniformity right: {smallest_uniformity_right}")

left_scalar = greatest_uniformity_left - smallest_uniformity_left
right_scalar = greatest_uniformity_right - smallest_uniformity_right

scaled_uniformity_left = [(u - smallest_uniformity_left) / left_scalar for u in uniformity_left] if left_scalar != 0 else [0] * len(uniformity_left)
scaled_uniformity_right = [(u - smallest_uniformity_right) / right_scalar for u in uniformity_right] if right_scalar != 0 else [0] * len(uniformity_right)
print(f"left scalar: {left_scalar}")
print(f"right scalar: {right_scalar}")
print(f"Scaled uniformity left: {scaled_uniformity_left}")
print(f"Scaled uniformity right: {scaled_uniformity_right}")
```
